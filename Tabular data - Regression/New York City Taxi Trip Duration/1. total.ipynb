```python
df = pd.read_csv("../input/nyc-taxi-trip-duration/train.zip")
```

```python
df.head()
```

지역 이미지를 기반으로 하는 데이터다. 따라서 혹시나 범위를 넘어서는 값이 있다면 없애자.

```python
xlim = [-74.03, -73.77]
ylim = [40.63, 40.85]
df = df[(df.pickup_longitude > xlim[0]) & (df.pickup_longitude < xlim[1])]
df = df[(df.dropoff_longitude > xlim[0]) & (df.dropoff_longitude < xlim[1])]
df = df[(df.pickup_latitude > ylim[0]) & (df.pickup_latitude < ylim[1])]
df = df[(df.dropoff_latitude > ylim[0]) & (df.dropoff_latitude < ylim[1])]
```

```python
loc_df = pd.DataFrame()
loc_df["longitude"] = list(df.pickup_longitude) + list(df.dropoff_longitude)
loc_df["latitude"] = list(df.pickup_latitude) + list(df.dropoff_latitude)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(loc_df.longitude, loc_df.latitude, '.', alpha=0.4, markersize=0.05)
plt.show()
```

### Cluster

```python
kmeans = KMeans(n_clusters=15, random_state=2, n_init=10)
kmeans.fit(loc_df)
loc_df["label"] = kmeans.labels_
```

```python
samples = loc_df.sample(20000)
plt.figure(figsize=(10, 10))
for label in samples.label.unique():
    plt.plot(samples.longitude[samples.label == label],
             samples.latitude[samples.label == label], '.', alpha=0.3, markersize=0.3)
plt.title("Clusters of New York")
plt.show()
```

```python
_, ax = plt.subplots(figsize=(10, 10))
for label in samples.label.unique():
    ax.plot(samples.longitude[samples.label == label],
            samples.latitude[samples.label == label],
            '.', alpha=0.4, markersize=0.1, color="gray")
    ax.plot(kmeans.cluster_centers_[label, 0], kmeans.cluster_centers_[label, 1],
            '.', color='r')
    ax.annotate(label, (kmeans.cluster_centers_[label, 0],
                       kmeans.cluster_centers_[label, 1]), color='b', fontsize=20)
    
    ax.set_title("Cluster Centers")
    plt.show()
```
